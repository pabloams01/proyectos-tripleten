```python
#eliminar los valores perdidos de name, genre y year_of_release ya que son muy poocos valores. name solo tiene 2 valores perdidos, genre 2  y 
#year of release tiene 269 que siguen siendo muy pocos (1.6%)
df = df.dropna(subset = ['name','genre','year_of_release'])
df.info()
```

```python
# Crear columna con grupo de plataformas para analizar por compañías para simplificar su análisis. 
def fun_plataforma (x):
    if x in ["3DS","DS","GB","GBA","GC","N64","NES","SNES","Wii","WiiU"]:
        return "Nintendo"
    elif x in ["PC","PCFX"]:
        return "PC"
    elif x in ["PS","PS2","PS3","PS4","PSP","PSV"]:
        return "PlayStation"
    elif x in ["X360","XB","XOne"]:
        return "XBox"
    else:
        return "Other"

df["platform_group"] = df["platform"].apply(fun_plataforma) 
df.sample(5)
```

```python
# sustituir 'tbd' por valor perdido en user_score: #este paso es necesario hacerlo para poder cambiar a float

df['user_score'] = df['user_score'].replace('tbd', np.nan)
df['user_score'].unique()
```

```python
#convertir datos en tipos necesarios:
# user score debe ser flotante para que sea un número y poder hacer cálculos
df['user_score'] = df['user_score'].astype(float)
```

```python
#multiplicar user_score x 10 para que sean comparables con critic_score que está del 1 al 100
df['user_score'] = df['user_score'] * 10
df[['critic_score','user_score']].describe().round()
```

```python
#cambiar year of release a entero porque los años no deben ser flotantes, son números enteros. 
df['year_of_release'] = df['year_of_release'].astype(int)
df.info()
```

```python
# Hacer más semantico el rating #que sea más entendible
def fun_rating (x):
    if x == "E" or x == "K-A":
        return "Everyone"
    elif x == "T":
        return "Teens"
    elif x == "M":
        return "Mature"
    elif x == "E10+":
        return "Everyone from 10 year"
    elif x == "EC":
        return "Early Childhood"
    elif x == "RP":
        return "Pending"
    elif x == "AO":
        return "Adults Only"
    else:
        return "Pending"

df["rating"] = df["rating"].apply(fun_rating)
df["rating"].value_counts()
```

```python
#crear columna total_sales
df['total_sales'] = df[['na_sales','eu_sales','jp_sales','other_sales']].sum(axis=1)
df.sample(5)
```

```python
#cuántos valores duplicados hay?
df.duplicated().sum()
```

```python
#¿cuánto tardan generalmente las nuevas plataformas en aparecer y desaparecer?
# Obtener el primer y último año de cada plataforma
platform_lifecycle = df.groupby('platform')['year_of_release'].agg(['min', 'max']).reset_index()
platform_lifecycle.columns = ['platform', 'first_year', 'last_year']

# Calcular la duración de vida de cada plataforma
platform_lifecycle['lifespan'] = platform_lifecycle['last_year'] - platform_lifecycle['first_year'] + 1

print(platform_lifecycle.sort_values('first_year'))
```

```python
# Calcular estadísticas del ciclo de vida
print("=== ESTADÍSTICAS DEL CICLO DE VIDA DE PLATAFORMAS ===")
print(f"Duración promedio de vida: {platform_lifecycle['lifespan'].mean():.1f} años")
print(f"Duración mediana: {platform_lifecycle['lifespan'].median():.1f} años")
print(f"Duración mínima: {platform_lifecycle['lifespan'].min()} años")
print(f"Duración máxima: {platform_lifecycle['lifespan'].max()} años")
print(f"Desviación estándar: {platform_lifecycle['lifespan'].std():.1f} años")
```

#Paso 5. Prueba de hipótesis 

Para el DataFrame aplicaremos las siguientes pruebas de hipótesis 

Hipótesis nula: Las calificaciones promedio de los usuarios para las plataformas XBox One y Pc son las mismas.
Hipótesis alternativa: Las calificaciones promedio de los usuarios para las plataformas Xbox One y pc son diferentes. 

Hipótesis nula: las calificaciones promedio de los usuarios para los géneros de Acción y Deportes no son diferentes.
Hipótesis alternativa: Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes. 

La formulación de las hipótesis nula se realiza siempre para probar que no hay diferencia y la hipótesis alternativa siempre trata de probar que los valores de las medias muestrales son estadísticamente distintas. 

La prueba a utilizar para ambas pruebas de hipótesis será la prueba t de student para muestras independientes ya que las medias muestrales a analizar son distintas.

Para ambas pruebas de hipótesis utilizaremos el valor de umbral alfa del 5%

```python

df_with_scores['platform'].unique()
```

```python

#prueba de hipótesis: las calificaciones promedio de los usuarios para las plataformas XBox One y Pc son las mismas.

#el espacio muestral se realizará sobre el dataframe con valores de calificaciones de usuarios 
xbox_scores = df_with_scores[df_with_scores['platform']== 'XOne']['user_score']
pc_scores = df_with_scores[df_with_scores['platform']=='PC']['user_score']

#Cálculo de varianza
xbox_var = xbox_scores.var()
pc_var = pc_scores.var()
# desviación estándar ara las plataformas Xbox One y PC

xbox_std = xbox_scores.std()
pc_std = pc_scores.std()

print(f'La varianza de Xbox One: {xbox_var}')
print(f"Desviación estándar Xbox One: {xbox_std:.3f}")
print(f'La varianza de PC: {pc_var}')
print(f"Desviación estándar PC: {pc_std:.3f}")


#calcular ratio para definir si la varianza debe considerarse como similar.
ratio = max(xbox_var, pc_var) / min(xbox_var, pc_var)
print(f'El radio de varianza es: {ratio}')




```

```python


#el radio de varianza es muy bajo, por lo que trataremos el valor como similar. 


alpha = .05


scores_result = st.ttest_ind(xbox_scores,pc_scores, equal_var=True)
print('Valor p: ', {scores_result.pvalue})

if scores_result.pvalue < alpha:
    print('Rechazamos la hipótesis nula')
else:
    print('No podemos rechazar la hipótesis nula')
```

Parece indicar que hay evidencia estadística para rechazar la hipótesis nula, por lo cual rechazamos las calificaciones promedio de los usuarios para las plataformas XBox One y Pc son las mismas. Esto significa que es existe una probabilidad significativa de que las calificaciones  de los usuarios de los juegos de XBox One y PC sean distintas.

```python
#prueba de hipótesis: las calificaciones promedio de los usuarios para los géneros de Acción y Deportes no son diferentes.

#extraer la muestra de la calificación de usuarios para ambos géneros 

scores_action = df_with_scores[df_with_scores['genre']=='Action']['user_score']
scores_sports = df_with_scores[df_with_scores['genre']=='Sports']['user_score']

#cálculo de varianza para las plataformas Xbox One y Pc
action_var = scores_action.var()
sports_var = scores_sports.var()

#cálculo de desviación estándar:
action_std = scores_action.std()
sports_std = scores_sports.std()

print(f'La varianza de Action: {action_var}')
print(f"Desviación estándar Action: {action_std:.3f}")
print(f'La varianza de Sports: {sports_var}')
print(f"Desviación estándar sports: {sports_std:.3f}")



#calcular ratio para definir si la varianza debe considerarse como similar.
ratio_genre = max(action_var, sports_var) / min(action_var, sports_var)
print(f'El radio de varianza es: {ratio_genre}')

```

```python
#prueba t 
genre_results = st.ttest_ind(scores_action,scores_sports, equal_var = True)
alpha = .05
print('Valor p: ', {genre_results.pvalue})
if genre_results.pvalue < alpha:
    print('Rechazamos la hipótesis nula')

else:
    print('No podemos rechazar la hipótesis nula')
```

El valor p es mayor mucho mayor que el nivel de significancia, por lo cual no tenemos evidencia estadística para inferir que los valores de las calificaciones de los juegos de acción y deportes sean significativamente distintos.
