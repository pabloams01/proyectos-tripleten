## Prueba las hipótesis estadísticas


Hipótesis nula = Los ingresos promedio entre los planes Ultimate y Surf son iguales (no hay diferencia)

Hipótesis alternativa = Los ingresos promedio entre los planes Ultimate y Surf son diferentes

el tipo de prueba será una de t de student de tipo independiente. con un nivel de confianza del 5%

```python
# Prueba las hipótesis

#extraer datos de ingresos por plan
surf_ingresos = df_complete[df_complete['plan']== 'surf']['costo_total']
ultimate_ingresos = df_complete[df_complete['plan']=='ultimate']['costo_total']

varianza_surf = surf_ingresos.var()
varianza_ultimate = ultimate_ingresos.var()


print(f"la varianza de la distribución del plan surf es: {varianza_surf}:2.f")
print(f"la varianza de la distribucón del plan ultimate es: {varianza_ultimate}:2.2f")
print(f"Ratio de varianzas: {varianza_surf/varianza_ultimate:.2f}")

```

```python
#hacer prueba t de student para muestras independientes
alpha = 0.05

#prueba t
results = st.ttest_ind(surf_ingresos,ultimate_ingresos, equal_var=False)
print('valor p:', results.pvalue)

if results.pvalue < alpha: #comparamos el valor con el umbral
    print('Rechazamos la hipótesis nula')
else:
    print('No podemos rechazar la hipótesis nula')

```
